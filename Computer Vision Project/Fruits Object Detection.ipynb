```python
def extract_text(filename):
    tree = et.parse(filename)
    root = tree.getroot()

    # extract filename
    image_name = root.find('filename').text
    # width and height of the image
    width = root.find('size').find('width').text
    height = root.find('size').find('height').text
    objs = root.findall('object')
    parser = []
    for obj in objs:
        name = obj.find('name').text
        bndbox = obj.find('bndbox')
        xmin = bndbox.find('xmin').text
        xmax = bndbox.find('xmax').text
        ymin = bndbox.find('ymin').text
        ymax = bndbox.find('ymax').text
        parser.append([image_name, width, height, name,xmin,xmax,ymin,ymax])

    return parser
```

```python
parser_train = list(map(extract_text,train_xmlfiles))
parser_test = list(map(extract_text,test_xmlfiles))
```

```python
parser_train
```

```python
parser_test
```

```python
data_train = reduce(lambda x, y : x+y,parser_train)
data_test = reduce(lambda x, y : x+y,parser_test)
```

```python
df_train = pd.DataFrame(data_train,columns = ['filename','width','height','name','xmin','xmax','ymin','ymax'])
df_test = pd.DataFrame(data_test,columns = ['filename','width','height','name','xmin','xmax','ymin','ymax'])
```

```python
# type conversion
cols = ['width','height','xmin','xmax','ymin','ymax']
df_train[cols] = df_train[cols].astype(int)
df_test[cols] = df_test[cols].astype(int)
df_train.info()
```

```python
# center x, center y
def center (df):
  df['center_x'] = ((df['xmax']+df['xmin'])/2)/df['width']
  df['center_y'] = ((df['ymax']+df['ymin'])/2)/df['height']
  # w
  df['w'] = (df['xmax']-df['xmin'])/df['width']
  # h
  df['h'] = (df['ymax']-df['ymin'])/df['height']
```

```python
center(df_train)
df_train
```

```python
center(df_test)
df_test
```

```python
# label encoding
def label_encoding(x):
    labels = {'orange':0, 'banana':1, 'apple':2}
    return labels[x]
```

```python
df_train['id'] = df_train['name'].apply(label_encoding)
df_test['id'] = df_test['name'].apply(label_encoding)
```

```python
import os
from shutil import move
```

```python
train = "/content/drive/MyDrive/Computer vision Project/Data/train"
test = "/content/drive/MyDrive/Computer vision Project/Data/test"


os.mkdir(train)
os.mkdir(test)
```

```python
cols = ['filename','id','center_x','center_y', 'w', 'h']
groupby_obj_train = df_train[cols].groupby('filename')
groupby_obj_test = df_test[cols].groupby('filename')
```

```python
#groupby_obj_train.get_group('000009.jpg').set_index('filename').to_csv('sample.txt',index=False,header=False)
# save each image in train/test folder and repective labels in .txt
def save_data(filename, folder_path, group_obj):
    # move image
    src = os.path.join("/content/drive/MyDrive/Computer vision Project/Data/train_zip/train",filename)
    dst = os.path.join(folder_path,filename)
    move(src,dst) # move image to the destination folder

    # save the labels
    text_filename = os.path.join(folder_path,
                                 os.path.splitext(filename)[0]+'.txt')
    group_obj.get_group(filename).set_index('filename').to_csv(text_filename,sep=' ',index=False,header=False)
```

```python
filename_series = pd.Series(groupby_obj_train.groups.keys())
```

```python
filename_series.apply(save_data,args=(train,groupby_obj_train))
```
